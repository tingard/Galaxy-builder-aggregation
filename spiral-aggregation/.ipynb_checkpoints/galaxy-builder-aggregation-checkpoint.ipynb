# Galaxy builder data aggregation

This notebook is a playground to develop the aggregation scripts for galaxy builder.

## Spiral arms

Want to cluster spiral arm points to identify individual arms, and then fit a smoothed spline to those points to provide arm number $m(r)$ and pitch-angle $\Psi(r)$ as functions of (deprojected) radius. Proposed steps as follows:

1) Group points (lines? Retains more data) into underlying drawn arm
2) fit spline

### Generation of clean data

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatProgress
from IPython.display import display
```

```python
def logSpiral(a, b, deltaTheta=0):
    return lambda t: np.transpose(np.concatenate((
        (a*np.cos(t+deltaTheta)*np.exp(b*t)).reshape(1, -1),
        (a*np.sin(t+deltaTheta)*np.exp(b*t)).reshape(1, -1)
    )))
```

```python
sFunc1 = logSpiral(1, 0.2)
sFunc2 = logSpiral(1, 0.2, deltaTheta=np.pi)
spiral1 = sFunc1(np.arange(0.1, 4*np.pi, 0.05))
spiral2 = sFunc2(np.arange(0.1, 4*np.pi, 0.05))
print(spiral1.shape)
plt.plot(spiral1[:, 0], spiral1[:, 1])
plt.plot(spiral2[:, 0], spiral2[:, 1])
```

Now we have our base spirals, we generate (slightly) noisy data by perturbing the vertices

```python
def makeNoisySpiral(s):
    return s + 0.5 * np.random.random(s.shape)

noisy1 = makeNoisySpiral(spiral1)
noisy2 = makeNoisySpiral(spiral2)

plt.plot(noisy1[:,0], noisy1[:,1])
plt.plot(noisy2[:,0], noisy2[:,1])
```

And now do that a bunch of times

```python
fakeClassifications = np.concatenate([
    np.concatenate((makeNoisySpiral(spiral1).reshape(1, -1), makeNoisySpiral(spiral2).reshape(1, -1))).reshape(1, 2, spiral2.shape[0], -1)
    for i in range(10)
])
print(fakeClassifications.shape)
# Data structure:
# Array of 
#     classifications = Array of 
#         annotations = Array of 
#             points = Array of 
#                 x = float
#                 y = float
print(fakeClassifications[0][0][0])
for c in fakeClassifications:
    color = None
    for arm in c:
        if color is None:
            p = plt.plot(arm[:, 0], arm[:, 1], alpha=0.3)
            color = p[0].get_color()
        else:
            p = plt.plot(arm[:, 0], arm[:, 1], alpha=0.3, color=color)
```

Okay, now we have something that could (maybe) represent a load of classifications. Time to move on to the grouping

## Classification grouping

We will make use of the `DBSCAN` clustering algorithm to try and group spirals


```python
# Imports
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler
```

Define a distance metric to identify how similar two arms are. The algorithm iterates through each arm and finds the nearest neighbour vertex of the partner arm. It then sums these nearest neighbours to provide a total "separation"

```python
X = fakeClassifications[0:2].reshape(-1, 250, 2)
a1 = X[0]
a2 = X[1]
a3 = X[2]

fig, ax = plt.subplots(ncols=2, figsize=(15, 7))

ax[0].plot(a1[:,0], a1[:,1])
ax[0].plot(a2[:,0], a2[:,1])

ax[1].plot(a1[:,0], a1[:,1])
ax[1].plot(a2[:,0], a2[:,1])

for p in a1:
    distances = np.sum((p-a2)**2, axis=1)
    minpoint = a2[np.argmin(distances)]
    plt.plot((p[0], minpoint[0]),(p[1], minpoint[1]), 'b--', alpha=0.1)
plt.suptitle('Figure illustrating the arc separation function')
```

#### Our distance function is thus defined:

```python
arcDistance = lambda a, b: sum(np.min(np.sum((p-a)**2, axis=1)) for p in b) + sum(np.min(np.sum((p-b)**2, axis=1)) for p in a)
```

#### A caveat

This function works well for arcs with a large number of points, but in cases where you are comparing one arc with lots of points to another with only a few, the discrepancies will add up!

```python
highRes = makeNoisySpiral(sFunc1(np.arange(0.1, 4*np.pi, 0.05)))
mediumRes = makeNoisySpiral(sFunc1(np.arange(0.1, 4*np.pi, 0.1)))
lowRes = makeNoisySpiral(sFunc1(np.arange(0.1, 4*np.pi, 0.5)))
plt.plot(lowRes[:, 0], lowRes[:, 1], alpha=0.6)
plt.plot(mediumRes[:, 0], mediumRes[:, 1], alpha=0.5)
plt.plot(highRes[:, 0], highRes[:, 1], alpha=0.4)
print('Measured distance:\nHigh to medium:', arcDistance(highRes, mediumRes), '\nHigh to low:', arcDistance(highRes, lowRes))
```

### Running the algorithm

`sklearn` `DBSCAN` requires that the input vector be two-dimensional, so we need to define a function which can map from the flattened arc coordinates to the distance calculating function `arcDistance`


```python
def aD(a, b):
    return arcDistance(a.reshape(-1, 2), b.reshape(-1, 2))
```

We can then initialise a `DBSCAN` fitter with a maximum neighbour distance of 100 (could be higher, will depend on spiral pitch angle / number of arms), and a minimum number of samples for a given clump to be called an arm (in this case 5). We also provide the metric.

Once the object is initialised we can start the fit! Extracted information is printed below.

```python
X = fakeClassifications.reshape(-1, 500)

# define "gold standard" labels
armLabelsTrue = (np.arange(len(X)) % 2)

# initialise fitter and fit! 
db = DBSCAN(eps=100, min_samples=5, metric=aD, n_jobs=-1).fit(X)

# Obtain clustering results
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
armLabels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(armLabels)) - (1 if -1 in armLabels else 0)

print('Estimated number of clusters: %d' % n_clusters_)
print("Homogeneity: %0.3f" % metrics.homogeneity_score(armLabelsTrue, armLabels))
print("Completeness: %0.3f" % metrics.completeness_score(armLabelsTrue, armLabels))
print("V-measure: %0.3f" % metrics.v_measure_score(armLabelsTrue, armLabels))
print("Adjusted Rand Index: %0.3f"
      % metrics.adjusted_rand_score(armLabelsTrue, armLabels))
print("Adjusted Mutual Information: %0.3f"
      % metrics.adjusted_mutual_info_score(armLabelsTrue, armLabels))
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(X, armLabels))
```

### YAY!

It works well! All of the sample arms are identified correctly, though they all have the same number of vertices and do not vary wildly from each other. The spirals they were generated from also have obvious separation.

The extracted spiral arms can be seen below.

```python
fig, ax = plt.subplots(ncols=len(np.unique(armLabels)), figsize=(15, 7))
for i, label in enumerate(np.unique(armLabels)):
    identified = X[np.where(armLabels == label)].reshape(-1, 2)
    ax[i].plot(identified[:, 0], identified[:, 1], 'b.', alpha=0.4)
```

## Recovering spiral information

We now have a point cloud representing the spiral. From here we could make use of the centre of the drawn disk / some other marked centre and fit a log spiral back to the data.

However this introduces biases - namely that the pitch angle is constant with radius (it isn't in general), the centre has been marked correctly and that we have correctly deprojected the galaxy (and that it isn't actually an ellipse).

For this reason I will instead make use of a smoothed spline fit, and then introduce the 2nd and 3rd biases again when I choose a center to measure pitch angle from!

```python
from scipy.interpolate import UnivariateSpline

def rThetaFromXY(x, y, mux=0, muy=0):
    return np.sqrt((x - mux)**2 + (y - muy)**2), np.arctan2((y - muy),(x - mux))

def xyFromRTheta(r, theta, mux=0, muy=0):
    return mux + r*np.cos(theta), muy + r*np.sin(theta)
```

Worth noting that $\theta \in [-\pi, \pi]$

The main problem encountered now is that of correctly ordering the data for a spline fit. Ideally the pointes would be ordered by theta, but as theta resets to $-\pi$ every full rotation this is nontrivial.

Again, clustering algorithms can come to our rescue! First, obtain ($r$, $\theta$)s for the points in each arm

```python
fig, ax = plt.subplots(ncols=len(np.unique(armLabels)), figsize=(15, 7))
armsRTheta = []
for i, label in enumerate(np.unique(armLabels)):
    identified = X[np.where(armLabels == label)].reshape(-1, 2)
    r, theta = rThetaFromXY(identified[:, 0], identified[:, 1])
    armsRTheta += [[r, theta]]
    # plot to demonstrate the resetting of theta
    ax[i].plot(theta, r, '.');
    ax[i].set_xlabel(r'$\theta$')
    ax[i].set_ylabel(r'$r$')
```

Now run a simple `DBSCAN` clusterer on the data points again

```python
fig, ax = plt.subplots(ncols=len(np.unique(armLabels)), figsize=(15, 7))
armsThetaLabels = []
for i, c in enumerate(armsRTheta):
    r, theta = c
    X2 = np.transpose(np.concatenate((theta.reshape(1, -1), r.reshape(1, -1))))
    dbThetas = DBSCAN(eps=1, min_samples=10, n_jobs=-1).fit(X2)
    
    print('Estimated number of clusters: %d' % n_clusters_)
    groupLabels = dbThetas.labels_
    armsThetaLabels += [groupLabels.copy()]
    for label in np.unique(groupLabels):
        cluster = X2[groupLabels == label]
        ax[i].plot(cluster[:, 0], cluster[:, 1], '.')
```

Now we have grouped our vertices, we can use their average radius to identify them in order and add $2\pi$ to each accordingly!

```python
fig, ax = plt.subplots(ncols=len(np.unique(armLabels)), figsize=(15, 7))
sortedPoints = []
for i, c in enumerate(armsRTheta):
    r, theta = c
    groupLabels = armsThetaLabels[i]
    meanRadius = [np.mean(r[groupLabels == label]) for label in np.unique(groupLabels)]
    
    dTheta = np.argsort(meanRadius) * 2.0 * np.pi
    unwrappedTheta = theta + dTheta[groupLabels]
    sortMask = np.argsort(unwrappedTheta)

    for label in np.unique(armsThetaLabels[i]):
        clusterR = r[groupLabels == label]
        clusterTheta = unwrappedTheta[groupLabels == label]
        ax[i].plot(clusterTheta, clusterR, '.', alpha=0.5, markersize=2)
    
    # sort by theta

    sortedPoints += [[unwrappedTheta[sortMask], r[sortMask]]]
    
```

### Spline fitting

We can now fit a spline to this unwound data!

```python
from scipy.interpolate import UnivariateSpline
fig, ax = plt.subplots(nrows=2, ncols=len(sortedPoints), figsize=(10, 7))
for i, (theta, r) in enumerate(sortedPoints):
    s = UnivariateSpline(theta, r, k=5)
    ax[0][i].plot(theta, r, '.', alpha=0.5, markersize=2)
    ax[0][i].plot(theta, s(theta), linewidth=1)
    ax[1][i].plot(*xyFromRTheta(r, theta))
    ax[1][i].plot(*xyFromRTheta(s(theta), theta))
```

And there we have it, a recovered "unbiased" spiral from noisy drawn lines!

This looks like a problem - with the fake data there is a distinct difference between the distances of arms in the same spiral, and arms in separate spirals. This distinction isn't there in the messier real classifications.

The problem probably lies with the arc separation function and the large discrepancies between point counts in the real classification. We'll try re-writing it to use a polygon distance-checking algorithm and taking the minimum of the two comparisons (a to b and b to a), rather than just looking for the nearest vertex.

```python
from numpy.linalg import norm
from matplotlib.patches import Ellipse

%load_ext line_profiler
```

The algorithm we'll implement is given by these functions:

```python
# function to get distance from a point (y) to a line connected by two vertices (p1, p2)
# from https://stackoverflow.com/questions/849211/shortest-distance-between-a-point-and-a-line-segment

def distance2(a, b):
    return (a[0]-b[0])**2 + (a[1]-b[1])**2
    
def dot(a, b):
    return a[0]*b[0] + a[1]*b[1]

def minimum_distance(p, v, w):
    # Return minimum distance between line segment vw and point p
    l2 = distance2(v, w);  # i.e. |w-v|^2 -  avoid a sqrt
    if l2 == 0.0: return np.sqrt(distance2(p, v));   # v == w case
    # Consider the line extending the segment, parameterized as v + t (w - v).
    # We find projection of point p onto the line. 
    # It falls where t = [(p-v) . (w-v)] / |w-v|^2
    # We clamp t from [0,1] to handle points outside the segment vw.
    t = max(0.0, min(1.0, dot(p - v, w - v) / l2))
    projection = v + t * (w - v)  # Projection falls on the segment
    return np.sqrt(distance2(p, projection))

def arcDistance(a, b):
    s1, s2 = 0, 0
    # add the minimum distance from each vertex to the other arm to a running total
    for point in a:
        s1 += min(minimum_distance(point, b[i], b[i+1]) for i in range(len(b)-1))
    # and for the other arm
    for point in b:
        s2 += min(minimum_distance(point, a[i], a[i+1]) for i in range(len(a)-1))
    # return the minimum distance 
    return (s1/len(a) + s2/len(b))

def aD2(a, b):
    return arcDistance(a.reshape(-1, 2), b.reshape(-1, 2))

# test by getting distance from (0,0) to [(1, -1), (1, 1)]
yCheck, p1Check, p2Check = np.array([0,0]), np.array([1, -1]), np.array([1, 1])
print('Passed test' if minimum_distance(yCheck, p1Check, p2Check) == 1.0 else 'Failed test :(')



# %lprun -f getDistance arcDistanceNew(noisy1, noisy2)  # detailed profiling
%timeit arcDistance(noisy1, noisy2)
```

It's a pretty painful running speed, can we speed it up using the magic of numpy..?

```python
# calculate dot of a(n,2), b(n,2): np.add.reduce(b1 * b2, axis=1)
# calucalte norm of a(n,2), b(n,2): np.add.reduce((a-b)**2, axis=1)
    
def vecDiffCalc(y, p1, p2, r):
    return np.sqrt((y[0]-p1[0])**2 + (y[1]-p1[1])**2 - r*((y[0]-p2[0])**2 + (y[1]-p2[1])**2))

def calcT(a):
    b1 = a[:, 0, :] - a[:, 1, :]
    b2 = a[:, 2, :] - a[:, 1, :]
    dots = np.add.reduce(b1 * b2, axis=1)
    l2 = np.add.reduce((a[:, 1] - a[:, 2])**2, axis=1)
    out = np.clip(dots / l2, 0, 1)
    return out

def getDiff(t, a):
    projection = a[:,1,:] + np.repeat(t.reshape(-1, 1), 2, axis=1) * (a[:,2,:] - a[:,1,:])
    out = np.add.reduce((a[:,0,:]-projection) * (a[:,0,:]-projection), axis=1)
    return np.sqrt(np.min(out))

vCalcT = np.vectorize(calcT, signature='(a,b,c)->(a)')
vGetDiff = np.vectorize(getDiff, signature='(a),(a,b,c)->()')

def minimum_distance2(a, b):
    d = 0
#     print(np.roll(b, 1, axis=0))
    m = np.zeros((a.shape[0], b.shape[0]-1, 3, 2))
    m[:, :, 0, :] = np.transpose(np.tile(a, [m.shape[1]+1, 1, 1]), axes=[1, 0, 2])[:, :-1, :]
    m[:, :, 1, :] = np.tile(b, [a.shape[0], 1, 1])[:, :-1, :]
    m[:, :, 2, :] = np.tile(np.roll(b, -1, axis=0), [a.shape[0], 1, 1])[:, :-1, :]
    t = vCalcT(np.array(m))
    return np.sum(vGetDiff(t, m))/a.shape[0]
    
def arcDistanceFast(a, b):
    return (
        minimum_distance2(a, b) +
        minimum_distance2(b, a)
    )

def aD2(a, b):
    return arcDistanceFast(a.reshape(-1, 2), b.reshape(-1, 2))

test = np.zeros((100, 2))
test[:, 0] += 1
test[:, 1] += np.arange(test.shape[0])
line = np.zeros((test.shape[0], 2))
line[:, 1] += np.arange(line.shape[0])

print('Is the same as original algorithm?')
assert (arcDistance(test, line) == arcDistanceFast(test, line))

%timeit minimum_distance2(noisy1, noisy2)
# %lprun -f calcT minimum_distance2(noisy1, noisy2)  # detailed profiling
```

### Another tack

Let's try grouping by points with a prior obtained by the mode number of spirals drawn by volunteers!

```python
from scipy.stats import mode
from sklearn.cluster import AgglomerativeClustering
from sklearn.neighbors import kneighbors_graph
```

We'll start by identifying the dominant (modal) number of drawn spiral arms

```python
modalN = mode([len(i) for i in spiralsWithLengthCut], axis=None)
m = modalN.mode[0]
print('Most likely spiral m={}, with {} classifications'.format(
    m, modalN.count[0]
))
```

Then use a connected Agglomerative clustering algorithm to try and group things

```python
model = AgglomerativeClustering(linkage='average',
                                connectivity=True,
                                n_clusters=m)
pointCloud = np.array([p for c in spiralsWithLengthCut for a in c for p in a]) - [256, 256]
```

```python
knn_graph = kneighbors_graph(pointCloud, 20, include_self=False)

model = AgglomerativeClustering(linkage='complete',
                                connectivity=knn_graph,
                                n_clusters=m)
model.fit(pointCloud)
```

Now let's have a look at the resulting labels. We'd naively expect points to be evenly distributed among the spiral arms...

```python
# expect an even distribution of labels
h = plt.hist(model.labels_, bins=np.arange(-1.5, np.max(model.labels_)+1.5), align='mid')
print('Distribution of points:', h[0])
plt.xticks(np.arange(-1, np.max(model.labels_)+1))
plt.yscale('log')
plt.ylabel('Number of points in group')
plt.xlabel('Identified groups')
```
